```python
meancountIa=np.mean(counts,axis=0)
meansqcountIa=np.mean(counts**2,axis=0)
sdIa=np.sqrt(meansqcountIa-meancountIa**2)

horzIa=np.delete(bins[0],len(bins)-1)

fig, ax = plt.subplots(figsize=(10, 8))

ax.fill_between(horzIa, meancountIa+sdIa, meancountIa-sdIa,alpha=0.3)
ax.plot(horzIa,meancountIa)
ax.plot([4,5.25],[0.09,0.009],color="LightGray")

# plt.grid(True)
ax.axis([0, 10, 0.0001, 2])
plt.yscale('log')
plt.title('z=3 single plane')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Probability of Magnification')
ax.legend(loc='upper right')
plt.show()
fig.savefig('MF_IrtyshIIIc.pdf', dpi=300)
```
