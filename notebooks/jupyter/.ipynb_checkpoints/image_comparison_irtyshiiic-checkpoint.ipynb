```python
import sys
import numpy as np
import operator
from scipy import optimize

sample = open('imagedata_python.dat', 'w') 
  


mult = np.genfromtxt('images_combined_multx.dat', dtype=({'names': ['rem','m','multdiff'], 'formats': ['i','i', 'i']}))

M = len(mult['m'])
#print(mult['m'],mult['rem'])
reimg = np.genfromtxt('reconstructed_images.dat', dtype=({'names': ['source','image','x', 'y'], 'formats': ['i','i', 'f8', 'f8']}))
#print(reimg['x'])
img = np.genfromtxt('images_combined.out', dtype=({'names': ['num','x', 'y', 'z'], 'formats': ['i', 'f8', 'f8', 'f8']}))
N = len(img['x']) #Number of images
#print(N)

m=0
n=0
image = np.zeros((2,M,9))
reimage = np.zeros((2,M,9))
for i in range(M):
    for j in range(mult['m'][i]):
        image[0,i,j], image[1,i,j] = img['x'][m], img['y'][m]-0.025
        m+=1
        
    for k in range(mult['rem'][i]):    
        reimage[0,i,k], reimage[1,i,k] = reimg['x'][n], reimg['y'][n]
        n+=1
# print(img['x'])
#print(reimage)
#print(m,n,"\n")
for i in range(M):
    dist = 0
    
    for ij in range(mult['m'][i]):
        dmin = 100
        sol = np.zeros(2)
        for ik in range(mult['rem'][i]):
            dist = (image[0,i,ij]-reimage[0,i,ik])**2 + (image[1,i,ij]-reimage[1,i,ik])**2
            if(dist<dmin): 
                dmin=dist 
                sol = reimage[0,i,ik],reimage[1,i,ik]
                #print(dmin)
        if(dmin**0.5<1): print('%3i %2i %2i %10.6lf %10.6lf %10.6lf %10.6lf %10.6lf'%(i+1,ij+1,mult['multdiff'][i],image[0,i,ij],image[1,i,ij],sol[0],sol[1],6*dmin**(0.5)),file=sample)
sample.close()
    
        #for k in range(mult['rem'][i]):
           
            

            
            


        
        
```
