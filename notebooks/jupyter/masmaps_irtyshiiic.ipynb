```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Rectangle

q=int((len(masswide['x']))**0.5)
print(q)
matplotlib.rcParams.update({'font.size': 10})

x = np.linspace(-240, 240, q)
y = np.linspace(-240, 240, q)
Xw, Yw = np.meshgrid(x, y)
Ztruewide=((masswide['kappa']*3.84562/3.13924)).reshape(q,q)

q=int((len(kappa['x']))**0.5)
print(q)
x = np.linspace(-120, 120, q)
y = np.linspace(-120, 120, q)
X, Y = np.meshgrid(x, y)
Zreconst=(kappa['kappa']).reshape(q,q)
Ztr=((mass['kappa']*3.84562/3.13924)).reshape(q,q)
Ztrue=np.transpose(Ztr)
Zrecon=np.transpose(Zreconst)
Zdiff=Zrecon-Ztrue
Zfdiff=(Zrecon-Ztrue)/Ztrue

fig = plt.figure(figsize=(10,10))
# plt.style.use('seaborn-white')
matplotlib.figure.SubplotParams(wspace=None, hspace=None)

ax = fig.add_subplot(2, 2, 1)
im=ax.pcolormesh(Xw,Yw,Ztruewide,cmap='plasma',alpha=1,vmax=5)
ax.contour(Xw,Yw,Ztruewide,[1],colors='white',alpha=0.3,linewidth=0.5)
# ax.scatter(img['x'],img['y'],s=10,alpha=0.5,color='white',facecolors='none') 
fig.colorbar(im, ax=ax, shrink=0.5)
r = Rectangle((-14*6, -14*6), 30*6, 30*6, edgecolor='white', facecolor='none',linewidth=1,alpha=0.4)
ax.add_patch(r)

ax.set_aspect('equal')
ax.set_title('True Mass')

ax = fig.add_subplot(2, 2, 2, projection='3d')

im=ax.plot_surface(Xw,Yw,Ztruewide, cmap='plasma',vmax=5)
fig.colorbar(im, ax=ax, shrink=0.5)
# ax.set_aspect('equal')
ax.set_title('3D True Mass ')
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))




ax = fig.add_subplot(2, 2, 3)

im=ax.pcolormesh(X,Y,Zrecon,cmap='plasma',alpha=1,vmax=5)
ax.contour(X,Y,Zrecon,[1],colors='white',alpha=0.3,linewidth=0.5)
ax.scatter(img['x']*6,img['y']*6,alpha=0.7,color='white',marker='+',linewidth=1,s=5) 
fig.colorbar(im, ax=ax, shrink=0.5)
ax.set_aspect('equal')
ax.set_title('Reconstructed Mass')
# r = Rectangle((-14, -14), 30, 30, edgecolor='white', facecolor='none',linewidth=0.5,alpha=0.4)
# ax.add_patch(r)
ax.axis([-14*6,16*6,-14*6,16*6])




ax = fig.add_subplot(2, 2, 4)

im=ax.pcolormesh(X,Y,Zfdiff,cmap='coolwarm',alpha=1)
ax.contour(X,Y,Zfdiff,[0],colors='white',alpha=0.3,width=0.5)
# ax.scatter(img['x'],img['y'],s=10,alpha=0.5,color='white',facecolors='none') 
# ax.scatter(img['x'],img['y'],alpha=0.7,color='black',marker='+',linewidth=1,s=5) 

fig.colorbar(im, ax=ax, shrink=0.5)
ax.set_aspect('equal')
ax.set_title('Fractional Mass Difference')
# r = Rectangle((-14, -14), 30, 30, edgecolor='white', facecolor='none',linewidth=1,alpha=0.4)
# ax.add_patch(r)
ax.axis([-14*6,16*6,-14*6,16*6])









plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=0.5)
# fig.savefig('/Users/agnivaghosh/Downloads/Lensing/mag/plots/Massmap_IrtyshIIIc.png', 
#             dpi=160, transparent=True, bbox_inches='tight')
```

```python
'''
=================
3D wireframe plot
=================

A very basic demonstration of a wireframe plot.
'''

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)

# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
