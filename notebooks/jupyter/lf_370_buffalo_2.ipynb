```python
def phiL(M):
    return 0.4*math.log(10)*phistar*10**(-0.4*(M-Mstar)*(alpha+1))*math.exp(-10**(-0.4*(M-Mstar)))
```

```python
M = np.linspace(-24, -14, num=100)
import math
phistar=10**(-3.88)
Mstar=-20.35
alpha=-1.98
x=np.zeros(len(M))
for i in range(len(M)): 
    x[i]=phiL(M[i])
plt.plot(M,x,'r--')
plt.yscale('log')
plt.axis([-24,-14,10**(-7),0.1])
```

```python
N=len(counts[1])
```

```python
fig, ax = plt.subplots()


phi=np.zeros([40,len(M)])
for i in range(40):
    for j in range(len(M)):
        for k in range(N):
            phi[i,j]+=(1/bins[i,k+1])*phiL(M[j]+2.5*np.log10(bins[i,k+1]))*counts[i,k]/80000
    plt.plot(M,phi[i,:],alpha=0.1)
plt.plot(M,x,'b--')
plt.yscale('log')
ax.set_title('Abell 370 | z=9.0')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Luminosity Function ($\phi$)')
plt.axis([-24,-14,10**(-7),0.1])
plt.show()

    
fig.savefig('LF_a370_BUFFALO.pdf', dpi=300, transparent=True, bbox_inches='tight')
```

```python
meancount=np.mean(phi,axis=0)
meansqcount=np.mean(phi**2,axis=0)
sd=np.sqrt(meansqcount-meancount**2)
```

```python
fig, ax = plt.subplots(figsize=(7,6))

plt.fill_between(M, meancount+sd, meancount-sd,alpha=0.6)
plt.plot(M,x,'b--')

# plt.grid(True)
plt.axis()
plt.yscale('log')
ax.set_title('Abell 370 | z=9.0')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Luminosity Function ($\phi$)')
# plt.xscale('log')
# ax2.set_title('IrtyshII')
# ax2.set_xlabel('Fractional Mass Dispersion')
# ax.legend(loc='upper right')

plt.axis([-24,-14,10**(-7),0.1])
plt.show()

fig.savefig('LF_a370_BUFFALO.png', dpi=300, transparent=True, bbox_inches='tight')
```
